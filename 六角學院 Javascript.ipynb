```javascript
// let, const 特性介紹 - 在設計上不可犯的錯誤

// 第一點 : 向上提升

                    
    console.log(a); // console.log 顯示 undefined
    var a = 3;  // var a 本身會先向上提升到最面,成為 undefine, 並不會說找不到 a ,但一直到這個位置 var a 才會被賦值
    console.log(a);
   
    // let 和 const 則不同,在哪一行建立變數,在該行才會開始建立並賦值,不會向上提升  
    console.log(a); // 抓不到值,顯示 Uncaught
    let a =3;
    console.log(a);



// 第二點 : 同個區塊上不能重複命名
// {}


    // var 可在同個區塊上重複命名

    var a = 1;
    var a = 2;  // 還是會顯示 var = 2 出來

    // const, let 則不允許

    let a = 1;
    let a = 2; // 會顯示 'has already been declared' 



// 第三點 : let 和 const 不會成為全域變數, 出現在window裡面

    const a = 1;
    var b = 1;
    window // 不會有 a 但會有 b
```

### // 01 什麼是API?

- API 本身就是抽象的，但我自己解釋廣義 API 就是「只要是你使用別人包裝好的程式語法與功能，就可稱作 API」。

- [應用程式介面](https://zh.wikipedia.org/wiki/%E5%BA%94%E7%94%A8%E7%A8%8B%E5%BA%8F%E6%8E%A5%E5%8F%A3)

與網際網路相連的端系統提供了一個應用程式介面（英語：Application Programming Interface，縮寫：API；又稱為應用程式編程介面）是軟體系統不同組成部分銜接的約定。

API規定了執行在一個端系統上的軟體請求網際網路基礎設施向執行在另一個端系統上的特定目的地軟體交付資料的方式。比方說：Alice使用郵政服務向Bob發一封信，郵政服務要求Alice將信放進信封中；在信封的中央寫上Bob的全名、位址和郵政編碼；封上信封；在信封的右上角貼上郵票；最後將信封丟進電子信箱里；郵政服務有自己的「郵政服務API」或一套規則，Alice必須這麼遵循，郵政服務才能把信寄給Bob；同理，網際網路也有一個傳送資料的程式必須遵循的API，使網際網路向接收資料的程式交付資料。---來源於：[《電腦網路-自頂向下學習法》](https://www.tenlong.com.tw/products/9787111599715?list_name=srh)

API 為 應用程式介面（英語：application programming interface，縮寫作API在網頁開發中，通常是指開發者與另一端程式串接的方法（通常以文件呈現),而 jQuery 並非是 API,也不是 Framework（正確為 library 函式庫）只有他的文件才能稱作 API;cros 沒開表示不允許跨域，另一種方式是使用後端;AJAX 串接其中一種就是透過 API，並沒有不流行的意思.
